
Array Slicing: Accessing Subarrays

Just as we can use square brackets to access individual array elements, we can also use them to access subarrays with the slice notation, marked by the colon (:) character. The NumPy slicing syntax follows that of the standard Python list; to access a slice of an array x, use this:

x[start:stop:step]

If any of these are unspecified, they default to the values start=0, stop=size of dimension, step=1. We'll take a look at accessing sub-arrays in one dimension and in multiple dimensions.


```python
x = np.arange(10)
x
```

```python
x[:5]  # first five elements
```

```python
x[:5]  # elements after index five
```

```python
x[4:7]  # sub-array of index 4, 5, 6
```

```python
x[::2]  # every other element
```

```python
x[2::2]  # every other element starting at index 2
```

A potentially confusing case is when the step value is negative. In this case, the defaults for start and stop are swapped. This becomes a convenient way to reverse an array:

```python
x[::-1]  # all elements, reversed
```

```python
x[3::-2]  # reversed every other from index 3
```


Introducing UFuncs

For many types of operations, NumPy provides a convenient interface into just this kind of statically typed, compiled routine. This is known as a vectorized operation. This can be accomplished by simply performing an operation on the array, which will then be applied to each element. This vectorized approach is designed to push the loop into the compiled layer that underlies NumPy, leading to much faster execution.


```python

#Consider following loop based implementation to ind reciprocals for each element of an array
np.random.seed(0)

def compute_reciprocals(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1.0 / values[i]
    return output

big_array = np.random.randint(1, 100, size=1000000)
%timeit compute_reciprocals(big_array)
```

```python

#Same operatoin using UFuncs applying '/' over array elements
%timeit (1.0 / big_array)
```


     

Vectorized operations in NumPy are implemented via ufuncs, whose main purpose is to quickly execute repeated operations on values in NumPy arrays. UFuncs are extremely flexible – before we saw an operation between a scalar and an array, but we can also operate between two arrays as well as multidimentional arrays.

Computations using vectorization through ufuncs are nearly always more efficient than their counterpart implemented using Python loops, especially as the arrays grow in size. Any time you see such a loop in a Python script, you should consider whether it can be replaced with a vectorized expression.

Q4.Consider array object named Obj with dimentions 3*4*2 and you want to convert it to 4*6 shape exlpain how will you do it ? Show this using a code cell, taking an example of 3*4*2 np array of random int between 1 and 50.

Answer:-To convert an array object named Obj with dimensions 3x4x2 to a 4x6 shape, you can use the reshape() method in NumPy. The key is that the total number of elements in the array must remain the same before and after the reshape operation. In this case, 3 * 4 * 2 = 24 elements, and 4 * 6 = 24 elements, so it's a valid reshape.

 create an example array and demonstrate the reshaping in the code cell.



```python
import numpy as np

# Create a 3x4x2 NumPy array with random integers between 1 and 50
np.random.seed(42) # for reproducibility
Obj = np.random.randint(1, 51, size=(3, 4, 2))

print("Original Obj array:\n", Obj)
print("Original Obj shape:", Obj.shape)
print("Total elements in Obj:", Obj.size)

# Reshape Obj to a 4x6 array
Obj_reshaped = Obj.reshape(4, 6)

print("\nReshaped Obj array (4x6):\n", Obj_reshaped)
print("Reshaped Obj shape:", Obj_reshaped.shape)
print("Total elements in reshaped Obj:", Obj_reshaped.size)
```

Q7.If you were to filter sub array view from above Obj object and want to apply modification in original and/or sub obj change respective element in other how will you achieve it?

Answer:-= Since slices already create views, any modification to a filtered view obtained from such a slice will also modify the original array's corresponding elements. If you only want to modify the sub-array and not the original, you'd explicitly create a copy using .copy().

Let's demonstrate how modifying an element in a filtered view affects both the view and the original arrays.

The code executed successfully, and the results further illustrate the view behavior in NumPy.

After creating region_for_filter from Obj_reshaped (which itself was a reshape of Obj) and then modifying an element within region_for_filter (e.g., changing region_for_filter[0, 1] from 40 to 555):

The region_for_filter array clearly shows the updated value.
The Obj_reshaped array, from which region_for_filter was derived, also reflects this change at the corresponding position (Obj_reshaped[2, 1] is now 555).
The original Obj array also shows this modification in its corresponding elements.




```python
import numpy as np

# Ensure Obj and Obj_reshaped are available from previous steps
# For demonstration, recreating with the last known state after previous modifications
Obj = np.array([[[ 39,  29],
        [ 15,  43],
        [   8,  21],
        [ 39,  19]],

       [[ 23,  11],
        [ 11,  24],
        [999,  40],
        [ 24,   3]],

       [[ 22,   2],
        [ 24,  44],
        [ 30,  38],
        [   2,  21]]])

Obj_reshaped = np.array([[ 39,  29,  15,  43,   8,  21],
       [ 39,  19,  23,  11,  11,  24],
       [999,  40,  24,   3,  22,   2],
       [ 24,  44,  30,  38,   2,  21]])

print("Initial Obj_reshaped:\n", Obj_reshaped)

# 1. Create a filtered sub-array view (using boolean indexing)
# Let's say we want to target elements in the last two rows and first three columns (similar to subObjCpy's area)
# and filter for elements greater than 30 within that region.

# Define the region of interest for filtering (last two rows, first three columns)
region_for_filter = Obj_reshaped[2:, :3]

# Create a boolean mask for filtering within this region
# For example, let's find elements in 'region_for_filter' that are greater than 30
mask = region_for_filter > 30

print("\nRegion for filter (last two rows, first three columns of Obj_reshaped):\n", region_for_filter)
print("\nBoolean mask (elements > 30 in the region):\n", mask)

# Apply the mask to get a filtered view. When using boolean indexing, the result is a copy if the indices are not consecutive or if the array is not writable directly.
# However, if we assign *back* to the boolean-indexed view, it modifies the original.
# Let's directly assign a new value to an element identified by the mask or a specific index within the region.

# Let's take the specific element at index [0,1] of 'region_for_filter' which is 40 and change it
# This element is Obj_reshaped[2, 1]
region_for_filter[0, 1] = 555 # Change the value 40 to 555

print("\nModified region_for_filter:\n", region_for_filter)
print("\nObj_reshaped after modifying region_for_filter:\n", Obj_reshaped)
print("\nObj after modifying region_for_filter (full array):\n", Obj)

# Conclusion will be provided in a separate markdown cell after execution
```
